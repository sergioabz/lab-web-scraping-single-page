```python
# 1. import libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd

# 2. Explore the dataset

songs_file = pd.read_csv('Spotify_Top_Hit_Playlist_2010_2022.csv')
songs_file.head()


# Dataset taken from: https://www.kaggle.com/datasets/josephinelsy/spotify-top-hit-playlist-2010-2022
```

```python
# Create a new DataFrame with columns "track_name" as Title an "artist_name" as Artist

not_hot_songs = pd.DataFrame().assign(Title=songs_file['track_name'], Artist=songs_file['artist_name'])
not_hot_songs.head()
```

```python
#Get DF ino

not_hot_songs.info()
```

```python
# Read the "hot_songs.csv" to merge the DataFrame and look for duplicates 

hot_songs = pd.read_csv('hot_songs.csv')
hot_songs.head()
```

```python
# Hot_songs shape

hot_songs.shape
```

```python
# Merge both DataFrame

duplicates = pd.concat([hot_songs, not_hot_songs], axis=1)
```

```python
# Look for missing values

duplicates.isna().sum()
```

```python
# Search for duplicate values between the two DataFrame's

duplicates1 = hot_songs.merge(not_hot_songs, how='inner', on='Title')
duplicates1
```

We have found two duplicate song + artist titles (Cruel Summer and Anti-Hero from Taylor Swift), we will remove them from the DataFrame "not_hot_songs".

```python
# Drop the two duplicate songs

not_hot_songs1 = not_hot_songs.copy().set_index('Title')
```

```python
# Drop the two duplicate songs

not_hot_songs1 = not_hot_songs1.drop(["Cruel Summer", "Anti-Hero"])
```
